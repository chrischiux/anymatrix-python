```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.table import Table

matrix_ID = ["matlab/wilkinson", "matlab/spiral",  "matlab/magic", "matlab/invhilb", "matlab/hilb", "matlab/hadamard", "matlab/compan"]
matrix_size = [8, 128, 1024, 4096]
comb = list(itertools.product(matrix_ID, matrix_size))
counter = 0

results = []

for args in comb:
    python_result = am.anymatrix(args[0], args[1])
    matlab_result = eng.anymatrix(args[0], float(args[1]))

    converted = np.array(matlab_result)
    
    result = np.array_equal(python_result, converted)
    if result == False:
        results.append((args[0], args[1], "Not Equal"))
    else:
        results.append((args[0], args[1], "Equal"))
        counter += 1

print(counter)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, len(results) * 0.5))

# Create a table
table = Table(ax, bbox=[0, 0, 1, 1])

# Add cells
nrows = len(results) + 1
ncols = 3
width, height = 1.0 / ncols, 1.0 / nrows

# Add header
table.add_cell(0, 0, width, height, text="Matrix ID", loc='center', facecolor='lightgrey')
table.add_cell(0, 1, width, height, text="Matrix Size", loc='center', facecolor='lightgrey')
table.add_cell(0, 2, width, height, text="Result", loc='center', facecolor='lightgrey')

# Add data
for i, (matrix_id, matrix_size, result) in enumerate(results):
    table.add_cell(i + 1, 0, width, height, text=matrix_id, loc='center', facecolor='white')
    table.add_cell(i + 1, 1, width, height, text=str(matrix_size), loc='center', facecolor='white')
    table.add_cell(i + 1, 2, width, height, text=result, loc='center', facecolor='lightgreen' if result == "Equal" else 'lightcoral')

# Add the table to the axis
ax.add_table(table)

# Hide the axes
ax.set_axis_off()

# Show the plot
plt.show()
```
